# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

## Plotting from list

- points in lisr are plotted as y-coordinates
- x-axis is from `0 to (N-1)` , N = number of elements in list

```python
plt.plot([1,2,3,4])
```

## Specifying values for both axes
> plot for y = x<sup>2</sup>

```python
x = range(5) # sequence of values for x - axes : [0,1,2,3,4]
plt.plot(x, [x1**2 for x1 in x])
```

```python
x = np.arange(0,5,0.01)
plt.plot(x, [x1**2 for x1 in x])
```

## Plotting multiple lines

```python
x = np.arange(0,5,0.01)
plt.plot(x,[x1 for x1 in x],
         x,[x1**2 for x1 in x],
         x,[x1**3 for x1 in x])
```

## Grid

- `grid()` takes in 2 boolean parameters : `True` or `False`

```python
plt.grid(True)
plt.plot(x,[x1 for x1 in x],
         x,[x1*2 for x1 in x],
         x,[x1*4 for x1 in x])
```

## Limiting the axes

- scale of the plot can be set using `axis(xmin, xmax, ymin, ymax)`

```python
plt.plot(x, [x1 for x1 in x],
         x, [x1*2 for x1 in x],
         x, [x1*4 for x1 in x])
plt.grid(True)
plt.axis([-1, 5, -1, 10])
```

- can also be done using `xlim()` and `ylim()`

```python
x = range(5)
plt.plot(x, [x1 for x1 in x],
         x, [x1*2 for x1 in x],
         x, [x1*4 for x1 in x])
plt.grid(True)
plt.xlim(-1, 5)
plt.ylim(-1, 10)
```

## Adding Labels

- can be done using `plt.xlabel('')` and `plt.ylabel('')`

```python
x = range(5)
plt.plot(x, [x1 for x1 in x],
         x, [x1*2 for x1 in x],
         x, [x1*4 for x1 in x])
plt.grid(True)
plt.xlabel('X-Axes')
plt.ylabel('Y-Axes')
```

## Adding title

- can be done using `plt.title('')`

```python
x = range(5)
plt.plot(x, [x1 for x1 in x],
         x, [x1*2 for x1 in x],
         x, [x1*4 for x1 in x])
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Polynomial Graph")
```

## Adding legend

- `label = ''` and `plt.label()`

```python
x = np.arange(5)
plt.plot(x, x, label = 'linear')
plt.plot(x, x*x, label = 'square')
plt.plot(x, x*x*x, label = 'cube')
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Polynomial Graph")
plt.legend()
```

## Adding markers

```python
x = [1, 2, 3, 4, 5, 6]
y = [11, 22, 33, 44, 55, 66]
plt.plot(x, y, 'bo')
for i in range(len(x)):
    x_cord = x[i]
    y_cord = y[i]
    plt.text(x_cord, y_cord, (x_cord, y_cord), fontsize = 10)
```

## Diiferent types of plot

### 1.Histogram

```python
y = np.random.randn(100, 100) # 100x100 array of a Gaussian distribution
plt.hist(y) # Function to plot the histogram takes the dataset as the parameter
```

```python
y = np.random.randn(1000)
plt.hist(y, 100)
plt.show()
```

### 2. Bar Chart

```python
# Default width of each bar is 0.8 units
# [1,2,3] Mid-point of the lower face of every bar
# [1,4,9] Heights of the successive bars in the plot
plt.bar([1,2,3], [1,4,9])
plt.show()
```

```python
dictionary = {'A':25, 'B':70, 'C':55, 'D':90}
for i, key in enumerate(dictionary):
    plt.bar(i, dictionary[key]) # Each key-value pair is plotted individually as dictionaries are not iterable
plt.show()
```

```python
dictionary = {'A':25, 'B':70, 'C':55, 'D':90}
for i, key in enumerate(dictionary):
    plt.bar(i, dictionary[key])
plt.xticks(np.arange(len(dictionary)), dictionary.keys()) # Adds the keys as labels on the x-axis
```

### 3. Pie  Chart

```python
plt.figure(figsize = (3,3)) # Size of the plot in inches
x = [40, 20, 5] # Proportions of the sectors
labels = ['Bikes', 'Cars', 'Buses']
plt.pie(x, labels = labels)
plt.show()
```

### 4. Scatter Plots

```python
# Two Gaussion distribution plotted
x = np.random.rand(1000)
y = np.random.rand(1000)
plt.scatter(x, y)
plt.show()
```

## Adding colours

- b = Blue
- c = Cyan
- g = Green
- k = Black
- m = Magenta
- r = Red
- w = White
- y = Yellow

```python
y = np.arange(1, 3)
plt.plot(y, 'y') # Specifying line colours
plt.plot(y+5, 'm')
plt.plot(y+10, 'c')
plt.show()
```

## Line styling

- - Solid line
- -- Dashed line
- -. Dash-Dot line
- : Dotted Line

```python
y = np.arange(1, 100)
plt.plot(y, '--', y*5, '-.', y*10, ':')
plt.show()
```

## Specifying markers

```python
y = np.arange(1, 3, 0.2)
plt.plot(y, '*',
        y+0.5, 'o',
        y+1, 'D',
        y+2, '^',
        y+3, 's')
```
